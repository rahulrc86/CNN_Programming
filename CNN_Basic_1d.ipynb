```python
import torch
import torch.nn as nn
```

```python
# -------------------------------------------------
# 1️⃣ Create fake real-world dataset
# 5 signals (batch)
# 2 measurements per time step (channels)
# 25 time steps (length)
# -------------------------------------------------

x = torch.randn(5, 2, 25)

print("Input shape:", x.shape)  # [5, 2, 25]
```

```python

# -------------------------------------------------
# 2️⃣ Define a simple 1D CNN layer
# in_channels = 2  (because data has 2 measurements)
# out_channels = 4 (we want 4 pattern detectors)
# kernel_size = 3  (looks at 3 time steps at once)
# -------------------------------------------------

conv = nn.Conv1d(
    in_channels=2,
    out_channels=4,
    kernel_size=3
)

print("Weight shape:", conv.weight.shape)
# Should be [4, 2, 3]


# -------------------------------------------------
# 3️⃣ Forward pass
# -------------------------------------------------

output = conv(x)

print("Output shape:", output.shape)
```
